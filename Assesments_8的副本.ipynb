```
new_words <- c("部務會報","權利歸屬證明書","時常","名義人","土地增值稅","農育權","典權","宗教財(社)團法人","少子化","高齡化","法制作業","陳報","送請")
writeLines(new_words,"new_words.txt")

stop_words <- c("在","之","更","個","未","而","也","的","與","從","中","人","已",
"仍","能","有","為","對","症","無","破","以","或","等","於","後","將","讓","時","間"
,"條","前","是","經","即","並","又","則","該","除","外","下")
writeLines(stop_words, "stop_words.txt")

cutter <- worker(user = "new_words.txt", stop_word ="stop_words.txt",
bylines = FALSE )
seg_words <- cutter[content]
seg_words
```

```
txt_freq <- freq(seg_words)
txt_freq <- arrange(txt_freq, desc(freq))
head(txt_freq)
```
