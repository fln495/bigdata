```
install.packages("SportsAnalytics")
install.packages("ggplot2")
```

```
library(SportsAnalytics)
library(ggplot2)
NBA1516 <- fetch_NBAPlayerStatistics("15-16")
str(NBA1516)
```

```
NBA1516Order <- NBA1516[order(NBA1516$TotalMinutesPlayed,decreasing = T),] ##總上場時間由高到低排列
NBA1516Order[,] ##取所有欄位
NBA1516Order[,"Name"] ##取名字的欄位
subset(NBA1516,Team=="GSW") ##搜尋隊伍欄位為"金州勇士隊"的球員
NBA1516[grepl("Curry",NBA1516$Name),] ##搜尋名字中有"Curry"的球員
```

```
ggplot(NBA1516,aes(Position, TotalPoints))+geom_point() ##橫軸為位置，縱軸為總得分數的散佈圖
```

```
ggplot(NBA1516,aes(Position))+geom_bar() ##橫軸為位置，縱軸為總數的長條圖
```

```
ggplot(NBA1516,aes(Position, TotalPoints))+geom_boxplot() ##橫軸為位置，縱軸為總得分數的盒狀圖
```

```
ggplot(NBA1516,aes(GamesPlayed, ThreesAttempted))+geom_line() ##橫軸為總場數，縱軸為三分出手數的折線圖
```
